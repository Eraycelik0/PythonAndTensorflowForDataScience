```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
npYasListesi = np.array(yasListesi)
npKiloListesi = np.array(kiloListesi)
```

```python
npKiloListesi
```

```python
npYasListesi
```

```python
plt.plot(npYasListesi,npKiloListesi,"m")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun yasa göre değişimi")
# plt.show()
```

```python
npDizisi1 = np.linspace(0,10,20)
```

```python
npDizisi1
```

```python
npDizisi2 = npDizisi1**3
npDizisi2
```

```python
plt.plot(npDizisi1,npDizisi2,"g--")
```

```python
plt.subplot(1,2,1)
plt.plot(npDizisi1,npDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şu an 2. grafiği çiziyorum
plt.plot(npDizisi2,npDizisi1,"b--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(npDizisi1,npDizisi2,"g")
figureAxes.set_xlabel("X ekseni ")
figureAxes.set_ylabel("Y ekseni ")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(npDizisi1,npDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik")
eksen2.plot(npDizisi2,npDizisi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows = 1, ncols = 2)

for eksen in benimEksenler:
    eksen.plot(npDizisi1, npDizisi2,"g")
    eksen.set_xlabel("X ekseni")
plt.tight_layout()
```

```python
yeniFigure = plt.figure()

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(npDizisi1, npDizisi1 ** 2, label="NumpyDizisi ** 2")
yeniEksen.plot(npDizisi1, npDizisi1 ** 3, label="NumpyDizisi ** 3")

yeniEksen.legend(loc=2)
```

```python
yeniFigure.savefig("benimfigur.png",dpi=200)
```
